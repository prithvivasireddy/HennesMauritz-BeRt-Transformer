```python
import pandas as pd
import json

# Read the data
articles_df = pd.read_csv("/Users/prithvi/Desktop/HnM_BerT/articles.csv", dtype=str)
customers_df = pd.read_csv("/Users/prithvi/Desktop/HnM_BerT/customers.csv", dtype=str)
```

```python
# Get column names for articles_df
article_columns = list(articles_df.columns)
print("Article columns:", article_columns)

# Get column names for customers_df
customer_columns = list(customers_df.columns)
print("Customer columns:", customer_columns)
```

```python



# Generate question-context-answer pairs
qa_pairs = []

# Article-related questions
for _, row in articles_df.iterrows():
    article_id = row['article_id']
    
    qa_pairs.append({
    "question": f"What is the product name of article ID {article_id}?",
    "context": f"Article ID {article_id} is {row['prod_name']}.",
    "answer": row['prod_name']
    })

    qa_pairs.append({
    "question": f"What is the product type name of article ID {article_id}?",
    "context": f"Article ID {article_id} has the product type {row['product_type_name']}.",
    "answer": row['product_type_name']
    })

    qa_pairs.append({
    "question": f"What is the product group name of article ID {article_id}?",
    "context": f"Article ID {article_id} belongs to the product group {row['product_group_name']}.",
    "answer": row['product_group_name']
    })

    qa_pairs.append({
    "question": f"What is the graphical appearance name of article ID {article_id}?",
    "context": f"Article ID {article_id} has a graphical appearance of {row['graphical_appearance_name']}.",
    "answer": row['graphical_appearance_name']
    })

    qa_pairs.append({
    "question": f"What is the color group name of article ID {article_id}?",
    "context": f"Article ID {article_id} has the color group {row['colour_group_name']}.",
    "answer": row['colour_group_name']
    })

    qa_pairs.append({
    "question": f"What is the department name of article ID {article_id}?",
    "context": f"Article ID {article_id} is in the department {row['department_name']}.",
    "answer": row['department_name']
    })

    qa_pairs.append({
    "question": f"What is the index code of article ID {article_id}?",
    "context": f"Article ID {article_id} has the index code {row['index_code']}.",
    "answer": row['index_code']
    })

    qa_pairs.append({
    "question": f"What is the index group name of article ID {article_id}?",
    "context": f"Article ID {article_id} is part of the index group {row['index_group_name']}.",
    "answer": row['index_group_name']
    })

    qa_pairs.append({
    "question": f"What is the section name of article ID {article_id}?",
    "context": f"Article ID {article_id} is in the section {row['section_name']}.",
    "answer": row['section_name']
    })

    qa_pairs.append({
    "question": f"What is the garment group name of article ID {article_id}?",
    "context": f"Article ID {article_id} belongs to the garment group {row['garment_group_name']}.",
    "answer": row['garment_group_name']
    })

    qa_pairs.append({
    "question": f"Can you provide a brief description of article ID {article_id}?",
    "context": f"Article ID {article_id} is described as follows: {row['detail_desc']}.",
    "answer": row['detail_desc']
    })



    # Customer-related questions
for _, row in customers_df.iterrows():
    customer_id = row['customer_id']
    qa_pairs.append({
    "question": f"What is the club member status of customer ID {customer_id}?",
    "context": f"Customer ID {customer_id} has a club member status of {row['club_member_status']}.",
    "answer": row['club_member_status']
    })

    qa_pairs.append({
    "question": f"How frequently does customer ID {customer_id} receive fashion news?",
    "context": f"Customer ID {customer_id} receives fashion news {row['fashion_news_frequency']}.",
    "answer": row['fashion_news_frequency']
    })

    qa_pairs.append({
    "question": f"What is the age of customer ID {customer_id}?",
    "context": f"Customer ID {customer_id} is {row['age']} years old.",
    "answer": row['age']
    })

    qa_pairs.append({
    "question": f"What is the postal code of customer ID {customer_id}?",
    "context": f"The postal code of customer ID {customer_id} is {row['postal_code']}.",
    "answer": row['postal_code']
    })
```

```python
# Save question-context-answer pairs as a JSON file
with open("/Users/prithvi/Desktop/HnM_BerT/qa_pairs.json", "w") as f:
    json.dump(qa_pairs, f, ensure_ascii=False, indent=2)
```
